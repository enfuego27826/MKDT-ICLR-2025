```python
import matplotlib.pyplot as plt
import torchvision
import torch
from dataloader import get_ssl_dataloaders
```

```python
def show_image_pair(x1, x2):
    def to_numpy(img_tensor):
        return img_tensor.clamp(0, 1).permute(1, 2, 0).numpy()

    fig, axs = plt.subplots(1, 2, figsize=(4, 4))
    axs[0].imshow(to_numpy(x1), interpolation='nearest')  # or 'bicubic'
    axs[0].set_title("Aug View 1")
    axs[0].axis("off")
    axs[1].imshow(to_numpy(x2), interpolation='nearest')
    axs[1].set_title("Aug View 2")
    axs[1].axis("off")
    plt.tight_layout()
    plt.show()
```
